```python
import pandas as pd
import folium
from google.colab import drive

drive.mount('/content/drive')
```

```python


file_path = '/content/drive/MyDrive/Big Data Engineering/2024/Final/Wine_Locations.csv'


df = pd.read_csv(file_path)


df.head()


map_center = [36.7783, -119.4179]  # Coordinates for California
map = folium.Map(location=map_center, zoom_start=6)

# Loop through the DataFrame and plot markers for each wine farm
for idx, row in df.iterrows():
    lat = row['latitude']
    lon = row['longitude']
    if pd.notna(lat) and pd.notna(lon):
        folium.Marker(location=[lat, lon], popup=row['Company Name']).add_to(map)


output_map_path = '/content/drive/MyDrive/Big Data Engineering/2024/Final/wine_farm_map.html'
map.save(output_map_path)

# Display the map in Colab
map
```
