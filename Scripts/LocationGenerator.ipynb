```python

file_path = '/content/drive/MyDrive/Big Data Engineering/2024/Final/California_Wine_Resturants.csv'


df = pd.read_csv(file_path)

# Initialize the geolocator with a user agent
geolocator = Nominatim(user_agent="wine_farm_locator")

# Set up the geocode function with rate limiting
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)


def get_coordinates(name):
    try:
        location = geocode(f"{name}, California, USA")
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except Exception as e:
        print(f"Error fetching coordinates for {name}: {e}")
        return None, None


df['latitude'], df['longitude'] = zip(*df['Company Name'].apply(get_coordinates))


output_file_path = '/content/drive/MyDrive/Big Data Engineering/2024/Final/Wine_Locations.csv'
df.to_csv(output_file_path, index=False)

df.head(30)
```
